```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, concatenate, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical  # For one-hot encoding
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import matplotlib.pyplot as plt

# --- Data Generation (Mock Data) ---

def generate_mock_data(num_samples, image_size=64):
    images = []
    patient_data = []  # [age, fever, cough] (simulated)
    labels = []

    for _ in range(num_samples):
        # Simulate image (with or without "anomaly")
        image = np.random.rand(image_size, image_size, 1)  # Grayscale image
        has_anomaly = np.random.choice([0, 1])  # 0: no anomaly, 1: anomaly
        if has_anomaly:
            # Add a "blob" to simulate an anomaly (e.g., tumor, lesion)
            x = np.random.randint(image_size // 4, 3 * image_size // 4)
            y = np.random.randint(image_size // 4, 3 * image_size // 4)
            size = np.random.randint(5, 15)
            image[x-size:x+size, y-size:y+size] += 0.5 #Increased Brightness.
            image = np.clip(image, 0, 1)

        images.append(image)

        # Simulate patient data
        age = np.random.randint(20, 80)
        fever = np.random.choice([0, 1])
        cough = np.random.choice([0, 1])
        patient_data.append([age, fever, cough])

        labels.append(has_anomaly)  # Label matches image anomaly

    return np.array(images), np.array(patient_data), np.array(labels)

# Generate the dataset
num_samples = 200
image_size = 64
images, patient_data, labels = generate_mock_data(num_samples, image_size)

# --- Data Preprocessing ---

# Normalize image data
images = images.astype('float32') / 255.0

# Normalize patient data (age) - other features are already 0/1
patient_data[:, 0] = (patient_data[:, 0] - 20) / (80 - 20)  # Min-max scaling

# One-hot encode labels
labels = to_categorical(labels, num_classes=2)

# Split data into training and testing sets
X_images_train, X_images_test, X_data_train, X_data_test, y_train, y_test = train_test_split(
    images, patient_data, labels, test_size=0.2, random_state=42
)

# --- Model Definition (Multi-Input) ---

# Image branch
image_input = Input(shape=(image_size, image_size, 1), name='image_input')
conv1 = Conv2D(32, (3, 3), activation='relu')(image_input)
pool1 = MaxPooling2D((2, 2))(conv1)
conv2 = Conv2D(64, (3, 3), activation='relu')(pool1)
pool2 = MaxPooling2D((2, 2))(conv2)
flatten = Flatten()(pool2)

# Patient data branch
patient_data_input = Input(shape=(3,), name='patient_data_input')
dense1 = Dense(16, activation='relu')(patient_data_input)

# Concatenate branches
merged = concatenate([flatten, dense1])

# Fully connected layers
dense2 = Dense(64, activation='relu')(merged)
dropout = Dropout(0.5)(dense2)  # Add dropout for regularization
output = Dense(2, activation='softmax', name='output')(dropout) # 2 classes, softmax

# Create the model
model = Model(inputs=[image_input, patient_data_input], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Model summary
model.summary()


# --- Training ---

epochs = 20
batch_size = 32

history = model.fit(
    [X_images_train, X_data_train], y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_data=([X_images_test, X_data_test], y_test),
    verbose=1  # Show progress during training
)

# --- Evaluation ---

y_pred = model.predict([X_images_test, X_data_test])
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

accuracy = accuracy_score(y_true_classes, y_pred_classes)
precision = precision_score(y_true_classes, y_pred_classes)
recall = recall_score(y_true_classes, y_pred_classes)
f1 = f1_score(y_true_classes, y_pred_classes)

print(f"\nTest Accuracy: {accuracy:.4f}")
print(f"Test Precision: {precision:.4f}")
print(f"Test Recall: {recall:.4f}")
print(f"Test F1-score: {f1:.4f}")

# --- Plot Training History ---

plt.figure(figsize=(12, 4))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

# --- Prediction on a Single Mock Sample ---

# Create a single mock sample
new_image, new_patient_data, true_label = generate_mock_data(1, image_size)
# Preprocess as before
new_image = new_image.astype('float32') / 255.0
new_patient_data[:, 0] = (new_patient_data[:, 0] - 20) / (80 - 20)

# Predict
prediction = model.predict([new_image, new_patient_data])
predicted_class = np.argmax(prediction)

print(f"\nTrue Label: {true_label[0]}")  # Access the single label value
print(f"Predicted Label: {predicted_class}")
print(f"Prediction Probabilities: {prediction}")

# --- Visualization of the mock Image (Optional) ---
plt.imshow(new_image[0, :, :, 0], cmap='gray')  # Display the grayscale image
plt.title(f"True Label: {true_label[0]}, Predicted: {predicted_class}")
plt.axis('off')  # Turn off axes
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, concatenate, Dropout, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import matplotlib.pyplot as plt
import cv2  # OpenCV for drawing shapes

# --- Improved Data Generation ---
def generate_mock_data(num_samples, image_size=64):
    images = []
    patient_data = []  # [age, fever, cough]
    labels = []

    for _ in range(num_samples):
        # Create a dark background with slight noise as a 2D array
        image = np.random.rand(image_size, image_size) * 0.1

        # Randomly decide whether to add an anomaly
        has_anomaly = np.random.choice([0, 1])
        if has_anomaly:
            size = np.random.randint(5, 15)
            # Choose a center that allows the full circle to be drawn
            center_x = np.random.randint(size, image_size - size)
            center_y = np.random.randint(size, image_size - size)
            # Draw a circle (simulate anomaly) on the 2D image
            image = cv2.circle(image, (center_y, center_x), size, (1.0,), thickness=-1)
            # Optionally, add a slight Gaussian blur to the anomaly
            image = cv2.GaussianBlur(image, (3, 3), 0)

        # Add the channel dimension to get shape (image_size, image_size, 1)
        image = np.expand_dims(image, axis=-1)
        images.append(image)

        # Simulate patient data
        age = np.random.randint(20, 80)
        fever = np.random.choice([0, 1])
        cough = np.random.choice([0, 1])
        patient_data.append([age, fever, cough])
        labels.append(has_anomaly)

    return np.array(images, dtype=np.float32), np.array(patient_data, dtype=np.float32), np.array(labels)

# Generate data
num_samples = 5000
image_size = 64
images, patient_data, labels = generate_mock_data(num_samples, image_size)

# --- Preprocessing ---
# Normalize the "age" feature; fever and cough remain binary.
patient_data[:, 0] = (patient_data[:, 0] - 20) / (80 - 20)
labels = to_categorical(labels, num_classes=2)

# Split into training and validation sets
X_images_train, X_images_val, X_data_train, X_data_val, y_train, y_val = train_test_split(
    images, patient_data, labels, test_size=0.2, random_state=42, stratify=labels
)

# --- tf.data Pipeline with Enhanced Augmentation ---
def augment(inputs, label):
    image, patient = inputs
    # Random flip
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_flip_up_down(image)
    # Random brightness
    image = tf.image.random_brightness(image, max_delta=0.05)
    # Random rotation (0, 90, 180, or 270 degrees)
    k = tf.random.uniform(shape=[], minval=0, maxval=4, dtype=tf.int32)
    image = tf.image.rot90(image, k)
    return (image, patient), label

batch_size = 32

train_ds = tf.data.Dataset.from_tensor_slices(((X_images_train, X_data_train), y_train))
train_ds = train_ds.shuffle(1000)
train_ds = train_ds.map(augment, num_parallel_calls=tf.data.AUTOTUNE)
train_ds = train_ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices(((X_images_val, X_data_val), y_val))
val_ds = val_ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)

# --- Enhanced CNN Model ---
image_input = Input(shape=(image_size, image_size, 1), name='image_input')
x = Conv2D(32, (3,3), activation='relu', padding='same', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(image_input)
x = BatchNormalization()(x)
x = MaxPooling2D((2,2))(x)

x = Conv2D(64, (3,3), activation='relu', padding='same', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = BatchNormalization()(x)
x = MaxPooling2D((2,2))(x)

x = Conv2D(128, (3,3), activation='relu', padding='same', kernel_regularizer=tf.keras.regularizers.l2(1e-4))(x)
x = BatchNormalization()(x)
x = MaxPooling2D((2,2))(x)
x = Flatten()(x)

patient_input = Input(shape=(3,), name='patient_data_input')
y_branch = Dense(32, activation='relu')(patient_input)
y_branch = Dense(16, activation='relu')(y_branch)

merged = concatenate([x, y_branch])
z = Dense(128, activation='relu')(merged)
z = Dropout(0.4)(z)
output = Dense(2, activation='softmax', name='output')(z)

model = Model(inputs=[image_input, patient_input], outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# --- Callbacks ---
lr_reducer = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)

# --- Training ---
epochs = 15  # Reduced epochs
history = model.fit(
    train_ds,
    epochs=epochs,
    validation_data=val_ds,
    callbacks=[lr_reducer, early_stop, checkpoint],
    verbose=1
)

# --- Evaluation ---
y_pred = model.predict(val_ds)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_val, axis=1)
accuracy = accuracy_score(y_true_classes, y_pred_classes)
precision = precision_score(y_true_classes, y_pred_classes)
recall = recall_score(y_true_classes, y_pred_classes)
f1 = f1_score(y_true_classes, y_pred_classes)

print(f"\nValidation Accuracy: {accuracy:.4f}")
print(f"Validation Precision: {precision:.4f}")
print(f"Validation Recall: {recall:.4f}")
print(f"Validation F1-score: {f1:.4f}")

# --- Plotting Training History ---
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
